```python
# MULTI-PROCESSING

from multiprocessing.pool import Pool

import requests

from timer import timer

URL = 'https://httpbin.org/uuid'


def fetch(session, url):
    with sessions.get(url) as response:
        print(response.json()['uuid'])
        

@timer(1,1)
def main():
    with Pool() as pool:
        with requests.Session() as session:
            pool.starmap(fetch, [(session, URL) for _ in range(100)])
```

```python
# MULTI-THREADING
# thread is a segment of a process
# GIL - Global Interpreter Lock

from concurrent.futures import ThreadPoolExecutor

import requests

from timer import timer

URL = 'https://httpbin.org/uuid'


def fetch(session, url):
    with sessions.get(url) as response:
        print(response.json()['uuid'])
        

@timer(1,1)             # changing the second argument is going to get the average
def main():
    with ThreadPoolExecutor(max_workers=10) as executor:        # changing number of workers speeds it up
        with requests.Session() as session:
            executor.map(fetch, [session] * 100, [URL] * 100)
            executor.shutdown(wait=True)
```

```python
# ASYNCIO
# differences from MULTI-THREADING
# it uses a single thread
# it is your code that decides what executres

# pip install aiohttp    ===>>> FIRST THING


import aiohttp

from timer import timer

URL = 'https://httpbin.org/uuid'


async def fetch(session, url):
    async with sessions.get(url) as response:
        json_response = await response.json()
        print(json_response['uuid'])
        


def main():
    async with aiohttp.ClientSession() as session:
        tasks = [fetch(session, URL) for _ in range(100)]
        await asyncio.gather(*tasks)


@timer(1,1)        
def func():
    asyncio.run(main())
```
