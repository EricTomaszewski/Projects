```python
import httpx
from bs4 import BeautifulSoup as bs

page = 1
authors = []
titles = []
prices = []
formats = []

while page !=35:        # because that's how many pages there
    async with httpx.AsyncClient() as client:
        url = f"https://www.bookdepository.com/bestsellers?page={page}"
        response = await client.get(url)
        html = response.content
        soup = bs (html, "lxml")
        for p in soup.find_all("p", class_="author"):
            authors.append(p.get_text(strip=True))
        for q in soup.find_all("h3", class_="title"):
            titles.append(q.get_text(strip=True))
        for r in soup.find_all("span", class_="sale-price"):
            prices.append(r.get_text(strip=True))
        for s in soup.find_all("p", class_="format"):
            formats.append(s.get_text(strip=True))
        page += 1
        print(page)
```

```python
print(authors)
print(len(authors))
print("")
print(titles)
print(len(titles))
print("")
print(prices)
print(len(prices))
print("")
print(formats)
print(len(formats))
```

```python
import pandas as pd

df = pd.DataFrame(titles, columns=["titles"])
df["author"] = pd.Series(authors)
df["prices"] = pd.Series(prices)
df["formats"] = pd.Series(formats)
print(list(set(formats)))
print(len(list(set(formats))))
print(df.prices.isnull().sum())
df

df.to_csv("bestsellers.csv")        # export data
```
