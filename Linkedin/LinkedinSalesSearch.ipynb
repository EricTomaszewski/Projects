```python
from sqlalchemy import Column, Integer, String, create_engine, Date, Time
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker

Base = declarative_base()

class Fundraising(Base):
    __tablename__ = 'fundraising'
    id = Column(Integer, primary_key=True, autoincrement=True)
    amount = Column(String)
    country = Column(String)
    contact = Column(String)
    description = Column(String)
    post_date = Column(Date)
    post_time = Column(Time)
    date_created = Column(Date)
    time_created = Column(Time)                     #  time_created = Column(Time, >>> default=None) <<<


engine = create_engine('sqlite:///fundraising.db')
Base.metadata.create_all(engine)

Session = sessionmaker(bind=engine)
session = Session()



# Add a single item to the database
entry = Fundraising(amount="$19,000,000", country="India", contact="Sanjay", description="a residential affordable development")
session.add(entry)
session.commit()



# Amend the amount field in the last entry to $20,000,000
last_entry = session.query(Fundraising).order_by(Fundraising.id.desc()).first()
last_entry.amount = "$20,000,000"
session.commit()



# Add to the description of element id 1 "across 8 sites"
entry_1 = session.query(Fundraising).filter_by(id=1).first()
entry_1.description += " across 8 sites"
session.commit()



session.close()
```
