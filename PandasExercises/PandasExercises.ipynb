```python
# 1) Sord DataFrame based on another list

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["C", 3], ["D", 4]],
                  columns = ["Col_A", "Col_B"])

sort_list = ["C", "D", "B", "A"]


# Solution
s = pd.Series(range(len(sort_list)), index = sort_list)

df.sort_values("Col_A", key = lambda x: x.map(s))
```

```python
# 2) Insert a column at a specific location in a DataFrame

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["C", 3], ["D", 4]],
                  columns = ["Col_A", "Col_B"])

new_column = ["P", "Q", "R", "S"]
insert_position = 1


# Solution
df.insert(insert_position, "Col_C", new_column)
df
```

```python
# 3) Select columns based on the column's Data Type

import pandas as pd

df = pd.DataFrame([["A",1,True], ["B",2,False], ["C",3,False], ["D",4,True]],
                  columns = ["Col_A", "Col_B", "Col_C"])

data_type = "bool"


# Solution
output_df = df.select_dtypes(include = data_type)
output_df
```

```python
# 4) Count the number of Non_NaN cells for each column

import pandas as pd
import numpy as np

df = pd.DataFrame([["A", np.NaN], [np.NaN, 2], ["C", np.NaN], ["D", 4]],
                  columns = ["Col_A", "Col_B"])


# Solution
df.count()
```

```python
# 5) Split DataFrame into equal parts

import pandas as pd
import numpy as np

df = pd.DataFrame([["A", 1], ["B", 2], ["C", 3], ["D", 4]], 
                  columns=["col_A", "col_B"])

parts = 2


# Solution
out_df1, out_df2 = np.split(df, parts)
print(out_df1)
print("")
print(out_df2)
```

```python
# 6) Reverse DataFrame row-wise or column-wise

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["C", 3], ["D", 4]], 
                  columns=["col_A", "col_B"])


# Solution
col_reverse = df.loc[:, ::-1]
print(col_reverse)

print("")

row_reverse = df.loc[::-1]
print(row_reverse)
```

```python
# 7) Rearrange columns of a DataFrame

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["C", 3], ["D", 4]], 
                  columns=["col_A", "col_B"])

rearrange_order = [1, 0]


# Solution
ro = df.iloc[:, rearrange_order]
ro
```

```python
# 8) Get alternate rows of a DataFrame

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["C", 3], ["D", 4]], 
                  columns=["col_A", "col_B"])


# Solution
output_df = df.iloc[::2]
output_df
```

```python
# 9) Insert a row at an arbitrary position

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["C", 3], ["D", 4]], 
                  columns=["col_A", "col_B"])

insert_pos = 1
insert_row = ["P", 5]


# Solution
df.loc[insert_pos - 0.1] = insert_row
df = df.sort_index()
df = df.reset_index(drop = True)
df
```

```python
# 10) Apply function to every cell of DataFrame

import pandas as pd

df = pd.DataFrame([[1,5], [2,6], [3,7], [4,8]],
                  columns=["Col_A", "Col_B"])

def func(num):
    return num+1


# Solution
df.applymap(func)
```

```python
# 1) The cumulative sum of a column in DataFrame

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["C", 3], ["D", 4]], 
                  columns=["col_A", "col_B"])


# Solution
df["Col_C"] = df.col_B.cumsum()
df
```

```python
# 2) Assign Unique IDs to every Group

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["A", 3], ["D", 4]], 
                  columns=["col_A", "col_B"])


# Solution
df["group_num"] = df.groupby("col_A").grouper.group_info[0] + 1
df
```

```python
# 3) Check if a column has NaN values

import pandas as pd
import numpy as np

df = pd.DataFrame([["A", np.NaN], ["B", 2], ["C", np.NaN], ["D", 4]], 
                  columns=["Col_A", "Col_B"])


# Solution
Col_A_check = df.Col_A.hasnans
print(Col_A_check)

print("")

Col_B_check = df.Col_B.hasnans
print(Col_B_check)
```

```python
# 4) Append a list as a row to a DataFrame

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["C", 3], ["D", 4]], 
                  columns=["col_A", "col_B"])

new_row = ["E", 5]


# Solution
df.loc[df.shape[0]] = new_row
df
```

```python
# 5) Get the first row of every unique value in a column

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["A", 3], ["D", 4]], 
                  columns=["Col_A", "Col_B"])


# Solution
df.groupby("Col_A").first()
```

```python
# 6) Identify the source of each row in Pandas Merge

import pandas as pd

df1 = pd.DataFrame([["A", 1], ["B", 2]], 
                  columns=["Col_A", "Col_B"])

df2 = pd.DataFrame([["A", 3], ["C", 4]], 
                  columns=["Col_A", "Col_C"])


# Solution
df = pd.merge(df1, df2, how="left", indicator=True)
df
```

```python
# 7) Filter n-largest and n-smallest values from a DataFrame

import pandas as pd

df = pd.DataFrame([["A", 200], ["B", 400], ["C", 100], ["D", 300]], 
                  columns=["Col_A", "Col_B"])

k=2


# Solution
new_df = df.nlargest(n=k, columns="Col_B")
print(new_df)

print("")
new_df_2 = df.nsmallest(n=k, columns="Col_B")
new_df_2
```

```python
# 8) Map categorical data to unique integral values

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["A", 3], ["D", 4]], 
                  columns=["Col_A", "Col_B"])


# Solution
df["new_col"] = pd.factorize(df.Col_A)[0]
df
```

```python
# 9) Add prefix to every column name

import pandas as pd

df = pd.DataFrame([["A", 1], ["B", 2], ["C", 3], ["D", 4]], 
                  columns=["Col_A", "Col_B"])


# Solution
df.add_prefix("pre_")
```

```python
# 10) Convert categorical columns to one hot value

import pandas as pd

df = pd.DataFrame([["A"], ["B"], ["C"], ["A"]], 
                  columns=["Col_A"])


# Solution
new_df = pd.get_dummies(df.Col_A)
new_df
```
